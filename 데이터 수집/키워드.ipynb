```python
import pandas as pd
file_path_new = 'news_title_jyp_스트레이키즈.csv'
df_new = pd.read_csv(file_path_new)
```

```python
#마이너스
df_new['열애']=df_new.iloc[:, 10].apply(lambda x: 1 if '열애' in str(x) else 0)
df_new['학교 폭력']=df_new.iloc[:, 10].apply(lambda x: 1 if '학교 폭력' in str(x) else 0)
df_new['음주 운전']=df_new.iloc[:, 10].apply(lambda x: 1 if '음주 운전' in str(x) else 0)
df_new['표절']=df_new.iloc[:, 10].apply(lambda x: 1 if '표절' in str(x) else 0)
df_new['인성 논란']=df_new.iloc[:, 10].apply(lambda x: 1 if '인성 논란' in str(x) else 0)
df_new['갑질']=df_new.iloc[:, 10].apply(lambda x: 1 if '재계약 성사' in str(x) else 0)
#플러스
df_new['재계약 성사']=df_new.iloc[:, 10].apply(lambda x: 1 if '재계약 성사' in str(x) else 0)
```
